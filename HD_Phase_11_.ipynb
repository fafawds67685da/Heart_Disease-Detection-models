```python
def predict(x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12,x13,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b):
  z1= x1*w1
  z2= x2*w2
  z3= x3*w3

  z4= x4*w4
  z5= x5*w5
  z6= x6*w6

  z7= x7*w7
  z8= x8*w8
  z9= x9*w9

  z10= x10*w10
  z11= x11*w11
  z12= x12*w12

  z13= x13*w13


  return (z1+z2+z3+z4+z5+z6+z7+z8+z9+z10+z11+z12+z13+b)

def cost_func(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,target1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b):

  y2=predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)

  mse=(y2-target1)**2
  return n.mean(mse)
```

```python
def update(age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b,lr):

  dLdW=n.mean(-2*age1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW2=n.mean(-2*sex1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW3=n.mean(-2*cp1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))

  dLdW4=n.mean(-2*trestbps1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW5=n.mean(-2*chol1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW6=n.mean(-2*fbs1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))

  dLdW7=n.mean(-2*restecg1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW8=n.mean(-2*thalach1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW9=n.mean(-2*exang1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))

  dLdW10=n.mean(-2*oldpeak1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))

  dLdW11=n.mean(-2*slope1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW12=n.mean(-2*ca1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))
  dLdW13=n.mean(-2*thal1*(target1-predict(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))

  dLdb=n.mean(-2*(target1-predict(age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)))

  w1=w1-lr*dLdW
  w2=w2-lr*dLdW2
  w3=w3-lr*dLdW3

  w4=w4-lr*dLdW4
  w5=w5-lr*dLdW5
  w6=w6-lr*dLdW6

  w7=w7-lr*dLdW7
  w8=w8-lr*dLdW8
  w9=w9-lr*dLdW9

  w10=w10-lr*dLdW10

  w11=w11-lr*dLdW11
  w12=w12-lr*dLdW12
  w13=w13-lr*dLdW13
  b=b-lr*dLdb
  return w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b
```

```python
def train(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b,lr,tol=1e-3,n_epochs=100, verbose=False):
  weights1=[w1]
  weights2=[w2]
  weights3=[w3]

  weights4=[w4]
  weights5=[w5]
  weights6=[w6]

  weights7=[w7]
  weights8=[w8]
  weights9=[w9]

  weights10=[w10]
  weights11=[w11]
  weights12=[w12]
  weights13=[w13]

  biases=[b]
  costs=[]

  ct=1
  while True:
    cost=cost_func(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,target1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b)
    costs.append(cost)
    if len(costs)>1 and abs(costs[-2]-costs[-1]) < tol:
      break

    w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b=update(age1,sex1,cp1,trestbps1,chol1,fbs1,restecg1,thalach1,exang1,oldpeak1,slope1,ca1,thal1,w1,w2,w3,w4,w5,w6,w7,w8,w9,w10,w11,w12,w13,b,lr)

    weights1.append(w1)
    weights2.append(w2)
    weights3.append(w3)

    weights4.append(w4)
    weights5.append(w5)
    weights6.append(w6)

    weights7.append(w7)
    weights8.append(w8)
    weights9.append(w9)

    weights10.append(w10)

    weights11.append(w11)
    weights12.append(w12)
    weights13.append(w13)

    biases.append(b)

    if verbose and ct %   n_epochs ==0:
      print("After ",ct," :- W1= ",w1," :- W2= ",w2," :- W3= ",w3," b= ",b," cost= ",cost)
    ct+=1

  return weights1,weights2,weights3,weights4,weights5,weights6,weights7,weights8,weights9,weights10,weights11,weights12,weights13,biases,costs

```
